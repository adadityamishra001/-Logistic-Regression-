Model building
1.Separate input and output features
2.Go for train test split for model validation.

1.Separate input and output features

```python
x=bank_data.drop(labels='y',axis=1)
y=bank_data[['y']]
```

```python
x
```

```python
y
```

2.Train test split for model validation.

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=12)
```

```python
#Training data
x_train.shape,y_train.shape
```

```python
#Test data
x_test.shape,y_test.shape
```

Model Training

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression()# initalization
logistic_model.fit(x_train,y_train)# model traning
```

```python
#Model Testing
#### Training data
y_train_pred = logistic_model.predict(x_train) #Trained question paper
y_train_pred
```

```python
#Test data
y_test_pred = logistic_model.predict(x_test) #New question
y_test_pred
```

 Model Evaluation

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score,classification_report
```

```python
#Training data
accuracy_score(y_train,y_train_pred)
```

```python
confusion_matrix(y_train,y_train_pred)
```

```python
precision_score(y_train,y_train_pred)
```

```python
recall_score(y_train,y_train_pred)
```

```python
print(classification_report(y_train,y_train_pred))
```

```python
from sklearn.metrics import roc_curve,roc_auc_score
fpr, tpr, thresholds = roc_curve(y, logistic_model.predict_proba (x)[:,1])
auc = roc_auc_score(y_train,y_train_pred)
print(auc)
import matplotlib.pyplot as plt
plt.plot(fpr, tpr, color='red', label='logit model ( area = %0.2f)'%auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
```

```python
#Test data
accuracy_score(y_test,y_test_pred)
```

```python
 confusion_matrix(y_test,y_test_pred)
```

```python
precision_score(y_test,y_test_pred)
```

```python
recall_score(y_test,y_test_pred)
```

```python
 print(classification_report(y_test,y_test_pred))
```
